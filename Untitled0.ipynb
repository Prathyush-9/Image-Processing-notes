```python
import numpy as np
import pandas as pd
import cv2 as cv
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/drive')
```

```python
image1 = io.imread('drive/MyDrive/0003.png')
cv2_imshow(image1)
```

```python
image2 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
cv2_imshow(image2)
```

```python
gray_image = cv.cvtColor(image2, cv.COLOR_RGB2GRAY)
cv2_imshow(gray_image)
```

```python
print(image2.dtype)
print(image2.shape[0])
print(image2.shape[1])
print(image2.shape[2])
plt.hist(image2.ravel(),bins = 256,range=[0,256])
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([image2],[i],None,[256],[0,200])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
